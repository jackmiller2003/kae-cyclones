```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import animation

import scipy.io
mat = scipy.io.loadmat('/g/data/x77/jm0124/fluids/DATA/FLUIDS/CYLINDER_ALL.mat')
data = np.swapaxes(mat['UALL'], 0, 1)
data2 = np.swapaxes(mat['VALL'], 0, 1)
data3 = np.swapaxes(mat['VORTALL'], 0, 1)
```

```python
data3.shape
```

```python
import random
random.shuffle(data)
lbl_array = ['u', 'v', 'vort']

for i, d in enumerate([data, data2, data3]):
    
    train_data = d[:100]
    val_data = d[100:125]
    test_data = d[125:150]

    np.save(f'/g/data/x77/jm0124/fluids/train_{lbl_array[i]}.npy', train_data)
    np.save(f'/g/data/x77/jm0124/fluids/valid_{lbl_array[i]}.npy', val_data)
    np.save(f'/g/data/x77/jm0124/fluids/test_{lbl_array[i]}.npy', test_data)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import animation

plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

fig = plt.figure()
dimension = (5, 5)
data = np.random.rand(dimension[0], dimension[1])
sns.heatmap(data, vmax=.8)

def init():
    sns.heatmap(np.zeros(dimension), vmax=.8, cbar=False)

def animate(i):
    data = np.random.rand(dimension[0], dimension[1])
    sns.heatmap(data, vmax=.8, cbar=False)

anim = animation.FuncAnimation(fig, animate, init_func=init, frames=20, repeat=False)

plt.show()
```
