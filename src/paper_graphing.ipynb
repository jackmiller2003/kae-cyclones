# Graphing Results for Paper

## Cyclone Task

```python
import csv 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def comparison_graph_training(csv_file, name):
    df = pd.read_csv(f'/home/156/jm0124/kae-cyclones/results/{csv_file}')
    plt.figure(figsize=(6.5,4),dpi=150)
    plt.xlabel('Epochs')
    plt.ylabel('Mean Squared Error')
    plt.plot(df['no_init_or_penalty-cyclone - forward loss'].ewm(span = 10).mean(),
             label='None')
    plt.plot(df['Maximal eigen-loss without eigen-initialisation - forward loss'].ewm(span = 10).mean(), 
             label='Loss')
    plt.plot(df['spectral_init_uniform-cyclone - forward loss'].ewm(span = 10).mean(), 
             label='Initialisation')
    plt.plot(df['sepctral_eigen_init_with_2_to_2_uniform_distribution - forward loss'].ewm(span = 10).mean(), 
             label='Initialisation & Loss')
    plt.ylim(0.1,1)
    plt.xlim(0,50)
    plt.legend(loc=2, prop={'size': 6})
    plt.savefig(f'{name}.png')
```

```python
comparison_graph_training('wandb_export_2022-06-16T16_37_29.399+10_00.csv', 'cyclone-comparison-graph-of-methods')
```
