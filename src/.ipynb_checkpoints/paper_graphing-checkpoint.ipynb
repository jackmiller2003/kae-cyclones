```python
labels = ['gaussian_double_dist_init-pendulum', 'gaussian_dist_init-pendulum', 'uniform_dist_init-pendulum', 'no methods pendulum']
```

```python
def comparison_graph_training(csv_file, labels, name, val_or_loss):
    df = pd.read_csv(f'/home/156/jm0124/kae-cyclones/results/{csv_file}')
    plt.figure(figsize=(6,3),dpi=150)
    plt.xlabel('Epochs')
    plt.ylabel('Mean Squared Error')
    for label in labels:
        print(f'{label} - forward {val_or_loss}')
        plt.plot(df[f'{label} - forward {val_or_loss}'].ewm(span = 10).mean(),
             label=f'{label}')
    
    # plt.plot(df['gaussian_double_dist_init-pendulum - forward loss'].ewm(span = 10).mean(),
    #          label='None')
    # plt.plot(df['Maximal eigen-loss without eigen-initialisation - forward loss'].ewm(span = 10).mean(), 
    #          label='Loss')
    # plt.plot(df['spectral_init_uniform-cyclone - forward loss'].ewm(span = 10).mean(), 
    #          label='Initialisation')
    # plt.plot(df['sepctral_eigen_init_with_2_to_2_uniform_distribution - forward loss'].ewm(span = 10).mean(), 
    #          label='Initialisation & Loss')
    # plt.ylim(0.1,1)
    # plt.xlim(0,50)
    plt.yscale('log')
    plt.legend(loc=1, prop={'size': 6})
    plt.savefig(f'{name}.png')
```

```python
comparison_graph_training('diss0_pend_val.csv', labels, 'cyclone-comparison-graph-of-methods')
```
